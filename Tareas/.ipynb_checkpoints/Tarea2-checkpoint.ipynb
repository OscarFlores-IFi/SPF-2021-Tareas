- **Evitando puntos ...**
 
Suponemos que no puedes usar la funcin `map` o una compresin de listas. Puede estar atascado con el bucle for. El ejemplo de bucle for tiene otra ineficiencia. Tanto `newlist.append ` como `word.upper` son referencias de funciones que se reevalan cada vez a travs del ciclo. El bucle original puede ser reemplazado por:

````python
# Almacenando funciones en variables (eliminar "." en las iteraciones de los for)
t1 = time.time()

new_list = []
append_new = new_list.append
upper_new = str.upper
for letter in words:
    append_new(upper_new(letter))

t2 = time.time()

print(t2 - t1)
````

- **Variables locales.**

La aceleracin final disponible para nosotros para la versin que no es un `map` del bucle for, es usar variables locales siempre que sea posible. Si el ciclo anterior se convierte como una funcin, las variables append y upper se convierten en locales. Python accede a las variables locales de forma mucho ms eficiente que las variables globales.

````python
# Generacin de funcin para utilizar variables locales
def upper_v_locales(words):
    new_list2 = []
    append_new2 = new_list2.append
    upper_new2 = str.upper
    for letter in words:
        append_new2(upper_new2(letter))
    return new_list2

t1 = time.time()
upper_v_locales(words)
t2 = time.time()

speed.append(t2 - t1)
print(t2 - t1)
````

## Funciones lambda

Las funciones `lambda` o funciones annimas, son un tipo de funcin que introduce python en su estructura, en donde se usa frecuentemente en dos situaciones principales: (a) si la funcin es bastante simple, y (b) si solo se usar una vez. Las funciones de Lambda se pueden usar donde sea que se requieran objetos de funcin

Python utiliza dos formas distintas de crear funciones: usando la sintxis `def` y `lambda`. A continuacin se muestran dos ejemplos de los usos de estas dos sintaxis:

#### Ejemplo raiz cuadrada

````python
###### Usando la sintxis def
def sqrt1(x):
    return np.sqrt(x)

###### Usando la sintaxis lambda
sqrt_lambda = lambda x: np.sqrt(x)

print('Usando la sintxis def obtenemos:',sqrt1(4))
print('Usando la sintxis lambda obtenemos:',sqrt_lambda(4))
````

Normalmente, las funciones se crean para uno de dos propsitos: **(a) reducir la duplicacin de cdigo, o (b) modularizar el cdigo.**

Si tienes una funcin que es usada una nica vez, no necesitas darle un nombre a la funcin. Puede ser 'annima' y adems puede definirla justo en el lugar donde desea usarlo. Ah es donde `lambda` es til.

## Porqu el uso de la funcin lambda es en ocasiones confuso?

**Primero:** Lambda es confuso porque: el requisito de que una lambda **pueda tomar solo una expresin** plantea la pregunta: Qu es una expresin?. En Python, cul es la diferencia entre una expresin y una declaracin (statement)?. Una buena respuesta es que una <font color=red> expresin devuelve (o evala) un valor</font>, mientras que una declaracin no.

Una pregunta que resulta entonces, es **qu tipo de cosas puedo, y no puedo, poner en una funcin lambda?** Y la respuesta a esa pregunta es bsicamente
   - Si no devuelve un valor, no es una expresin y no se puede poner en una lambda.
   - Si puede imaginarlo en una declaracin de asignacin, en el lado derecho del signo igual, es una expresin y puede colocarse en una lambda.

Usando estas reglas podramos entonces llegar a las siguientes conclusiones:

*1*. Los enunciados de asignacin no se pueden usar en lambda. En Python, las declaraciones de asignacin no devuelven nada, ni siquiera Ninguno (nulo).
     
**Ejemplo**

````python
# Asignaciones en una funcin lambda
asignacion = lambda x,y: x == y
asignacion(6,7)
````

*2*. Cosas simples como operaciones matemticas, operaciones de cadenas, comprensin de listas, etc. estn bien en una lambda.

````python
# Ejemplo
########## Sumar elemento a elemento de una lista
suma = lambda a,b: a + b
l1 = [1 ,2, 3]
l2 = [4, 5, 6]
# 1. Usando for
sum1 = []
for i in range(len(l1)):
    sum1.append(suma(l1[i], l2[i]))

# 2. Usando compresin de listas
sum2 = [suma(l1[i], l2[i]) for i in range(len(l1))]
sum2

########## Obtener el mismo resultado usando vectores
sum3 = np.array(l1) + np.array(l2)
sum3
````

````python
a = print(3)
b = lambda x: print(x)
b(4)
````

*3*. Las llamadas a funciones son expresiones. Est bien colocar una llamada de funcin en un lambda y pasar argumentos a esa funcin. Al hacerlo, envuelve la llamada a la funcin (argumentos y todo) dentro de una nueva funcin annima. Incluso la funcin `print` puede ser usada dentro de una funcin lambda.

- **Ejemplo** (compresin de listas)

````python
palabra = 'hola grupo es la prueba'

# Poner print dentro de una lambda, que regrese la letra mayscula del argumento
funUpper = lambda word: print(word.upper())

# Poner dentro de una lambda una funcin que regrese la letra mayscula
funUpper2 = lambda word: word.upper()

a = funUpper2(palabra)
print(a)
````

- **Ejemplo** (funciones anidadas)

````python
def suma(a,b):
    return a+b

# funcin annima
suma_potencia = lambda n,a,b: \
                print('la suma de %d y %d a la potencia %d es: %.1f' % (a,b,n, suma(a,b)**n))
suma_potencia(10,30,4.2)
````

*4*. Las expresiones condicionales, son expresiones (y no simplemente una sintaxis diferente para una instruccin if / else). Devuelven un valor y se pueden usar en una lambda
   
**Ejemplo**

````python
# Probar estas dos funciones

# Escribir funcin annima que regrese True si x>y de lo contrario False
some_condition = lambda x, y: True if x > y else False

# Utilizar la funcin anterior como condicin en una lambda
f1 = lambda x, y: 100 if some_condition(x,y) else 0

# Crear una funcin que regrese 'big' si x>100 de lo contrario 'small'
f2 = lambda x: 'big' if x > 100 else 'small'

# Verificar f1 y f2 en la siguiente lista
l1 = [100,25,200,5]

# 1. Usando for
r_f2 = []
for num in l1:
    r_f2.append(f2(num))
# print(r_f2)

r_f1 = []
for i in range(len(l1) - 1):
    r_f1.append(f1(l1[i], l1[i + 1]))
print(r_f1)

# 2. Usando compresin de listas
compresion2 = [f2(num) for num in l1]
compresion1 = [f1(l1[i], l1[i + 1]) for  i in range(len(l1) - 1)]

print(compresion1)
````

**Segundo:** Lambda es confuso porque: lambda generalmente se describe como una herramienta para crear funciones, pero una especificacin lambda no contiene una declaracin de **return**.

La declaracin de *return* est, en cierto sentido, implcita en una lambda. Como una especificacin lambda debe contener solo una expresin, y esa expresin debe devolver un valor, una funcin annima creada por lambda devuelve implcitamente el valor devuelto por la expresin. Esto tiene perfecto sentido. An as, la falta de una declaracin de *return* explcita es, creo, parte de lo que hace que sea difcil asimilar lambda, o al menos, difcil asimilarlo rpidamente.

**Ejemplo de combinacin de los dos tipos de funciones**

````python
# Crear una funcin que regrese una funcin annima que sea un incremento de cierto valor n (entrada)
def make_incrementor(n):
    return lambda x: x + n

f = make_incrementor(22)
i1 = 0; i2 = 5

print('Incrementar en %d unidades'% i1, f(i1),'\nIncrementar en %d unidades' %i2, f(i2))
# # f(100)
# f(1)
````

> `r = map(funcin, secuencia)`

Con Python 3, map () devuelve un iterador, si queremos obtener los valores numricos debemos usar compresin de listas o el comando `list`.

**Ejemplos**

````python
Celsius = [39.2, 36.5, 37.3, 37.8]

# Convierta las temperaturas en C a frados fahrenheit --> eq = float(9)/5)*i + 32

# 1. Usando llenado de listas
f1 = []
for temp in Celsius:
    f1.append((float(9)/5)*temp + 32)
print(f1)

# 2. Usando funciones y compresin de listas
f_conversion = lambda t: (float(9)/5)*t + 32
f2 = [f_conversion(temp) for temp in Celsius]
print(f2)

# 3. Usando funciones y funciones map
f3 = list(map(f_conversion, Celsius))
print(f3)

# 4. Usando vectores
f4 = f_conversion(np.array(Celsius))
print(f4)

# Comparar
````

## - Funcin reduce()
La funcin
> `reduce(func, secuencia)`

continuamente aplica la funcin `func()` a la `secuencia`. Devuelve un solo valor.

Si `secuencia = [s1, s2, s3, ..., sn]`, la llamada de la funcin `reduce(func, secuencia)` funciona as: 
 
   - Al principio, los primeros dos elementos de secuencia se aplicarn a `func`, es decir, `func(s1, s2)`. La lista en la que se trabaja con reduce() se ve as: `[func (s1, s2), s3, ..., sn ] `
   - En el siguiente paso, se aplicar func sobre el resultado anterior y el tercer elemento de la lista, es decir, `func (func (s1, s2), s3) `
   - La lista se ve as ahora: `[func(func (s1, s2), s3), ..., sn] `

## - Funcin `deque()`
Devuelve un nuevo objeto deque inicializado de izquierda a derecha (usando `append()`) con datos de iterable. Si no se especifica iterable, la nueva deque es vaca.

Los deques son una generalizacin de pilas y colas (el nombre se pronuncia 'deck' y es la abreviatura de 'cola de dos extremos' ('double-ended queue')). Los Deques admiten 'memory efficient appends' , salto de elementos desde cualquier lado del deque con aproximadamente el mismo rendimiento $O(1)$ en cualquier direccin.

Aunque los objetos de lista admiten operaciones similares, estn optimizados para operaciones rpidas de longitud fija e incurren en costos de movimiento de memoria $O(n)$ para operaciones de insercin `pop(0)` e insercin `insert(0, v)` que cambian tanto el tamao como la posicin de la representacin de datos subyacentes .

Si no se especifica maxlen o es `None`, los deques pueden crecer hasta una longitud arbitraria. De lo contrario, la deque est limitada a la longitud mxima especificada. Una vez que se llena una etiqueta de longitud limitada, cuando se agregan nuevos elementos, se descarta el nmero correspondiente de elementos del extremo opuesto. Los deques de longitud limitada proporcionan una funcionalidad similar al filtro de cola en Unix. Tambin son tiles para rastrear transacciones y otros grupos de datos donde solo la actividad ms reciente es de inters.

````python
from collections import deque
d = deque('ghi')                 # make a new deque with three items
for elem in d:                   # iterate over the deque's elements
    print(elem.upper())
````

````python
d.append('j')                    # add a new entry to the right side
d.appendleft('f')                # add a new entry to the left side
d     
````

````python
# d.pop()                          # return and remove the rightmost item
d.popleft()                      # return and remove the leftmost item
list(d)  
````

````python
list(reversed(d))                # list the contents of a deque in reverse
````

````python
'h' in d                         # search the deque
````

````python
d.extend('jkl')                  # add multiple elements at once
d
````

````python
d.rotate(1)                      # right rotation
d
````

````python
d.rotate(-1)                     # left rotation
d
````

````python
deque(reversed(d))               # make a new deque in reverse order
````

````python
d.clear()                        # empty the deque

d.extendleft('abc')              # extendleft() reverses the input order
d
````

## Comparacin tiempo de simulacin para grandes cantidades de datos 

Lo que se pretende en este apartado es mostrar la forma mas eficiente de realizar clculos cuando se tiene gran cantidad de datos. Se comparn las tcnicas tales como: ciclos for, paquete numpy, compresin de listas y la funcin map. Para efectos de simulacin se desea comparar los tiempos necesarios para calular el valor del $log_{10}$ para un vector de tamao $1\times 10^7$, que contiene valores aleatorios flotantes obtenidos mediante un distribucin normal, para las cuatro tcnicas mencionadas anteriormente.

````python
# Nmero de muestras
import time
import math
N = 10**7
# Lista con valores aleatorios
l1 = np.random.normal(50,5,N)
speed = []

############ Usando ciclo for
t1 = time.time()
l2 = []
for item in l1:
    l2.append(math.log10(item))
t2 = time.time()
print("Con ciclos for y agregando los resutados toma {} segundos".format(t2-t1))
speed.append(t2-t1)
````

````python
########### Usando numpy
t1 = time.time()
a2 = np.log10(l1)
t2 = time.time()
print("Usando la librera Numpy y el mtodo log10 toma {} segundos".format(t2-t1))
speed.append(t2-t1)
````

````python
########### Usando comprensin de listas
t1 = time.time()
cl = [math.log10(i) for i in l1]
t2 = time.time()
print("Usando compresin de listas toma {} segundos".format(t2-t1))
speed.append(t2-t1)
````

````python
########### Usando map function 
t1 = time.time()
m1 = list(map(lambda x:math.log10(x),l1))
t2 = time.time()
print("Usando la funcin map para el mtodo log10 toma {} segundos".format(t2-t1))
speed.append(t2-t1)
````

````python
########### Usando deque function
t1 = time.time()
deq = deque(maxlen=N)
for item in l1:
    deq.append(math.log10(item))
t2 = time.time()

print("Usando deque function toma {} segundos".format(t2-t1))
speed.append(t2-t1)
````

````python
########### Usando pandas
import pandas as pd
pandas = pd.Series(l1)

t1 = time.time()
pandas = pandas.apply(math.log10)
t2 = time.time()

print("Usando pandas toma {} segundos".format(t2-t1))
speed.append(t2-t1)
# pandas
````

````python
ind = np.arange(1, 7)
plt.figure(figsize=[9,5])
plt.bar(ind,speed)
plt.xticks(ind,['Ciclo for', 'Numpy', 'Compresin de\n listas','Funcin map', 'Funcin deque', 'Pandas'])
plt.ylabel('Tiempo gastado ')
plt.show()
````

### Ejercicio: Vectorizar distribucin de poisson
$$\textbf{Funcin de densidad de probabilidad}\\p(k)=\frac{\lambda^k e^{-\lambda}}{k!},\quad k\in \mathbb{N}$$

````python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sps 
from itertools import cycle
````

````python
## Forma de resolver el problema de manera tpica (para olvidar)
cycol = cycle('bgrcmk')

p = lambda k,l: (l**k*np.exp(-l))/sps.gamma(k)

# Como se hara de forma ineficiente
l= 1

# Grficas en diferentes subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,10));

ax1.title.set_text('Densidad de distrucin de probabilidad')
ax2.title.set_text('Distrucin de probabilidad acumulada')

for i in range(3):
    P = []
    c = next(cycol)
    for k in range(60):
        P.append(p(k,l))
        print(k,l,p(k,l))
    
    ax1.plot(P,'o',c=c,label='$\lambda$=%s' % l, markersize=2)        
    ax1.legend()
    
    ax2.plot(np.cumsum(P),'o',c=c,label='$\lambda$=%s' % l, markersize=3)
    ax2.legend()
    ax2.hlines(np.cumsum(P),range(len(P)),range(1,len(P)+1),color=c)
    l +=20
    
plt.subplots_adjust(hspace=.4)
plt.show()
````

````python
import time

# Grficas en diferentes subplots
def plotting(P):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,10));

    ax1.title.set_text('Densidad de distrucin de probabilidad')
    ax2.title.set_text('Distrucin de probabilidad acumulada')
    
    cycol = cycle('bgr') # since cmk colors are not used, they can be deleted. 
    
    for i in range(len(P)):
        c = next(cycol)

        ax1.plot(P[i],'o',c=c,label='$\lambda$=%s' % l, markersize=2)        
        ax1.legend()

        ax2.plot(np.cumsum(P[i]),'o',c=c,label='$\lambda$=%s' % l, markersize=3)
        ax2.legend()
        ax2.hlines(np.cumsum(P[i]),range(len(P[i])),range(1,len(P[i])+1),color=c)

    plt.subplots_adjust(hspace=.4)
    plt.show()
````

````python
# Inicializar variables
N = 60

## Forma de resolver el problema de manera no tan tpica.
p = lambda k,l: (l**k*np.exp(-l))/sps.gamma(k)

l = [20*i + 1 for i in range(3)] # Como se hara de forma no tan ineficiente
k = [i for i in range(N)]

#l = np.arange(3)*20 +1 # Un poquito ms eficiente, pero se generaban resultados raros usandolos. 
#k = np.arange(N)
````

````python
# Evaluar funcion p

t0 = time.time()
## Preallocating: 
P = np.zeros((N, len(l)))

## Definiendo funcin y llamarla en compresin de listas. 
def modify_P(j,i):
    global P
    P[j,i] = p(k[j],l[i])
    #print(k[j],l[i])
    
[modify_P(j,i) for i in np.arange(len(l)) for j in range(len(k))]
P
print(time.time() - t0)
````

````python
plotting(P.T)
````

````python
# Evaluar funcin P ms 'eficiente'.

t0 = time.time()
## Preallocating: 
P = np.zeros((N, len(l)))

## Usando un for tpico
for i in np.arange(len(l)):
    for j in range(len(k)):
        P[j,i] = p(k[j],l[i])
P
print(time.time() - t0)
````

````python
plotting(P.T)
````
