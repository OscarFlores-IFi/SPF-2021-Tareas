```python
# Función de aceptación y rechazo usando compresión de listas
def Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
#     R1 = np.random.rand(N)
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') \
         for i in range(len(R1))] ;
    plt.show();
    
    x = [R1[i] for i in range(len(R1)) if condition[i]]
    
    ap = len(condition[condition==True])/len(condition)
    return x, ap
```

```python
N = 500  # número de puntos

# Función objetivo
x = np.arange(0,1,0.01)

plt.plot(x,f(x),'k')
plt.plot(x,t(x),'b')
R2 = np.random.rand(N)
R1 = np.random.rand(N)

_,ap = Acep_rechazo(R2,R1,f,t)

plt.show()
```

```python
# Función de aceptación y rechazo usando compresión de listas
def Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
#     R1 = np.random.rand(N)
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') \
         for i in range(len(R1))]            
    plt.show()
    
    x=[R1[i] for i in range(len(R1)) if condition[i]]
    ap = len(condition[condition==True])/len(condition)
    return x, ap
```

```python
# Validación del método
N = 500 # número de puntos a simular
# Como estoy tomando t(x) constante solo es necesario generar valores aleatorios U~(0,1)
x=np.arange(0,1,0.01)
R2 = np.random.rand(N)
R1 = np.random.rand(N)
# Prueba de función creada
plt.plot(x,f(x),'k')
plt.plot(x,t(x),'b')
_,aa=Acep_rechazo(R2,R1, f,t)
plt.show()
```
